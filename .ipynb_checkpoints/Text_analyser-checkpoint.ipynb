```python
import pandas as pd
import numpy as np
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

### List of positive and negative words

```python
positive = pd.read_csv('MasterDictionary/positive-words.txt', header=None)
negative = pd.read_csv('MasterDictionary/negative-words.txt', header=None, encoding='ISO-8859-1')
```

```python
positive.columns=['words']
negative.columns=['words']
```

### Counting number of sentences

```python
def number_of_sentences(text):
    z = text
    for i in z:
        text_str = ''.join(i)
        
    nltk.download('punkt')
    from nltk.tokenize import sent_tokenize

    sentences = text_str

    tokenize_sentences = sent_tokenize(sentences)
    number_of_sentences = len(tokenize_sentences)
    return number_of_sentences 
```

### Counting personal pronouns

```python
def count_personal_pronouns(text_str):   
    import re

    pronounRegex = re.compile(r'I|he|him|her|ours|it|me|she|them|they|us|we|you',re.I)
    pronouns = pronounRegex.findall(text_str)
    pronouns_count = len(pronouns)
    return pronouns_count
```

### Function to remove puncs

```python
def remove_punc(text):
    for ele in text:
        if ele in string.punctuation:
            text = text.replace(ele,'')
    return text   
```

### Average word length

```python
def avg_word_length(text):
    z = text
    for i in z:
        text_str = ''.join(i)
    char_count = len(text_str)
    words_without_puncs = remove_punc(text_str)
    word_counts = len(word_tokenize(words_without_puncs))
    
    return (char_count,word_counts)
```

### Function to remove stop word and tokenizing

```python
import os
all_files = os.listdir("StopWords/")

def read_first_line(file):
    f = open('StopWords/{}'.format(file),'r')
    StopWords_ = f.read().split('\n')
    return StopWords_

output_strings = list(map(read_first_line, all_files))
    
stop_list = list(np.concatenate(output_strings))
```

```python
def remove_stop_words(text):
    stop_words = set(stopwords.words('english'))
    stop_words.update(stop_list)
    
    word_tokens = word_tokenize(text)
    filter_sentence = []
    for w in word_tokens:
        if w.lower() not in stop_words:
            filter_sentence.append(w)
    return filter_sentence
```

### Skipping numeric values from text

```python
skip_numeric = lambda x : [w for w in x if not w.isnumeric()]
```

### Function for syllables word_count

```python
def syllables_word_count(text):
    
    from nltk.corpus import cmudict
    d = cmudict.dict()
    
    def syllables_count(word):
        return [len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]]
    
    words_with_syllables_list=[]
    complex_words_list = []
    for i in text:
        try:
            words_with_syllables_list.append(syllables_count(i))
        except Exception as e:
            pass

    for i in words_with_syllables_list:
        if i[0] > 2:
            complex_words_list.append(i)

    total_syllables = len(words_with_syllables_list)
    complex_words = len(complex_words_list)
    
    return (total_syllables,complex_words)
```

```python
POSITIVE SCORE	NEGATIVE SCORE	POLARITY SCORE	SUBJECTIVITY SCORE	AVG SENTENCE LENGTH	
PERCENTAGE OF COMPLEX WORDS	FOG INDEX	AVG NUMBER OF WORDS PER SENTENCE		
	SYLLABLE PER WORD	

COMPLEX WORD COUNT
WORD COUNT
PERSONAL PRONOUNS	
AVG WORD LENGTH
```

```python
def text_analyser(data):
    main_data = data
    data = data.map(remove_punc)
    data = data.map(remove_stop_words)
    data = data.map(skip_numeric)
    
    positive_word_count = data.map(lambda x: len([w for w in x if w in list(positive['words'])]))
    negative_word_count = data.map(lambda x: len([w for w in x if w in list(negative['words'])]))
    word_count = avg_word_length(main_data)[1]
    Polarity_score = (positive_word_count-negative_word_count)/((positive_word_count+negative_word_count)+ 0.000001)
    Subjectivity_Score = (positive_word_count[0] + negative_word_count[0])/ ((word_count) + 0.000001)
    avg_sentence_length = word_count/number_of_sentences(main_data)
    
    percentage_of_complex_word = syllables_word_count(main_data[0])[1]/word_count
    fog_index = 0.4*(avg_sentence_length+percentage_of_complex_word)
    avg_no_words_per_sentence = word_count/number_of_sentences(main_data)
    syllables_per_word = syllables_word_count(main_data[0])[0]/word_count
    
    average_word_len = avg_word_length(main_data)[0]/word_count
    personal_pronouns = count_personal_pronouns(main_data[0])
    complex_word_count = syllables_word_count(main_data[0])[1]
    
    final_dict = {'positive_score':positive_word_count[0], 'negative_score':negative_word_count[0],
                  'polarity_score':Polarity_score[0],'Subjectivity_Score':Subjectivity_Score,
                  'fog_index':fog_index,
                  'Avg_sentence_length':avg_sentence_length,'Percentage of complex words':percentage_of_complex_word,
                  'avg_no_of_words_per_sentence':avg_no_words_per_sentence,'word_count':word_count,
                  'syllables_per_word':syllables_per_word,'Complex_Word_count':complex_word_count,
                  'Personal pronouns':personal_pronouns,'Average word length':average_word_len}
    
    return final_dict
    
    
```
